```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Generate some simple data
# Let's use a linear relationship y = 2x + 1
x_train = np.array([0, 1, 2, 3, 4, 5], dtype=float)
y_train = np.array([1, 3, 5, 7, 9, 11], dtype=float)
```

```python
# Build the model
model = Sequential([
    Dense(units=1, input_shape=[1])  # Single neuron with one input feature
])
```

```python
# Compile the model
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# Train the model
model.fit(x_train, y_train, epochs=100)
```

```python
# Make predictions
print("Prediction for 6:", model.predict(np.array([6])))
```
